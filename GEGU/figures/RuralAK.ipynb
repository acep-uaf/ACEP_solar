```python
ambler_ira_new = add_time_col(ambler_ira)
ambler_water_new = add_time_col(ambler_water)
cchrc_sharp_new = add_time_col(cchrc_sharp)
cchrc_solar165_new = add_time_col(cchrc_solar165)
fort_yukon_new = add_time_col(fort_yukon)
koyukuk_new = add_time_col(koyukuk)
manley_new = add_time_col(manley)
nenana_new = add_time_col(nenana)
ruby_health_new = add_time_col(ruby_health)
ruby_W_new = add_time_col(ruby_W)
```

```python
air_annual_values,air_month = annual_Norm(ambler_ira_new)
water_annual_values,water_month = annual_Norm(ambler_water_new)
sharp_annual_values,sharp_month = annual_Norm(cchrc_sharp_new)
solar_annual_values,solar_month = annual_Norm(cchrc_solar165_new)
fort_annual_values,fort_month = annual_Norm(fort_yukon_new)
koyukuk_annual_values,koyukuk_month = annual_Norm(koyukuk_new)
manley_annual_values,manley_month = annual_Norm(manley_new)
nenana_annual_values,nenana_month = annual_Norm(nenana_new)
health_annual_values,health_month = annual_Norm(ruby_health_new)
w_annual_values,w_month = annual_Norm(ruby_W_new)
```

```python
new_air = pd.DataFrame({'Date':air_month,'ambler_ira':air_annual_values})
new_water = pd.DataFrame({'Date':water_month,'ambler_water':water_annual_values})
new_sharp = pd.DataFrame({'Date':sharp_month,'cchrc_sharp':sharp_annual_values})
new_solar = pd.DataFrame({'Date':solar_month,'cchrc_solar165':solar_annual_values})
new_fort = pd.DataFrame({'Date':fort_month,'fort_yukon':fort_annual_values})
new_koyukuk = pd.DataFrame({'Date':koyukuk_month,'koyukuk':koyukuk_annual_values})
new_manley = pd.DataFrame({'Date':manley_month,'manley':manley_annual_values})
new_nenana = pd.DataFrame({'Date':nenana_month,'nenana':nenana_annual_values})
new_health = pd.DataFrame({'Date':health_month,'ruby_health':health_annual_values})
new_w = pd.DataFrame({'Date':w_month,'ruby_W':w_annual_values})
```

```python
result = pd.merge(new_air,new_water,on='Date',how='outer')
result = pd.merge(result,new_sharp,on='Date',how='outer')
result = pd.merge(result,new_solar,on='Date',how='outer')
result = pd.merge(result,new_fort,on='Date',how='outer')
result = pd.merge(result,new_koyukuk,on='Date',how='outer')
result = pd.merge(result,new_manley,on='Date',how='outer')
result = pd.merge(result,new_nenana,on='Date',how='outer')
RuralAK = pd.merge(result,new_health,on='Date',how='outer')
# result = pd.merge(result,new_solar,on='Date',how='outer')
# result = pd.merge(result,new_fort,on='Date',how='outer')
```
